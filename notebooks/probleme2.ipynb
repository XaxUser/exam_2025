## Partie I : un exemple de PointCLoud data

Pour construire le jeu de données, on simule un terrain couvert de deux types de bâtiments : des immeubles de forme rectangulaire aux toits plats et des igloos (dômes).
```
batch_size = 6
input_points, target_list, target_points  = gen_pointcloud(batch_size)


for i in range(batch_size):
  print(i)
  # Représentation 3D des nuages de points et
  # les paramètres elev et azim permettent de changer l'angle de vue
  plot_triplets(input_points[i].transpose(0,1).cpu(),
                elev=75, azim=0)

  # Cibles : les points appartenant aux toitures d'igloos sont
  # dans la classe 1, les autres, dans la classe 0.
  plot_triplets(target_points[i].transpose(0,1).cpu(),
                title='Cibles',
                cbar_label='classe')

  # Note: target_points contient non seulement les classes
  # mais aussi les coordonnées x et y des points, pour
  # faciliter leur visualisation
```

**Q1** A quoi correspondent les différentes dimensions de *input_points* ?

La variable `input_points` est un ensemble de nuages de points, et chaque nuage de points est une collection de coordonnées 3D, c'est-à-dire de triplets $(x, y, z)$ pour chaque point. En supposant que chaque point est représenté par un vecteur de dimension 3, les dimensions de *input_points* devraient être :

1. **Dimension 0** : Le nombre d'exemples (ou de nuages de points) dans le batch. Ici, il est donné par `batch_size` qui vaut 6, donc il y a 6 nuages de points.
2. **Dimension 1** : Le nombre de points dans chaque nuage de points. Ce nombre peut varier en fonction de la taille des nuages de points générés dans la fonction `gen_pointcloud`. C’est généralement une constante dans le jeu de données ou générée dynamiquement.
3. **Dimension 2** : Les coordonnées 3D des points. Chaque point est représenté par ses coordonnées $ (x, y, z) $, donc cette dimension vaut 3.


**Q2** Les points d'un nuage sont-ils rangés dans un ordre particulier ?

Dans la plupart des cas, les nuages de points sont échantillonnés aléatoirement dans l'espace 3D. Cependant, le traitement des nuages de points dans des architectures comme PointNet est conçu pour être invariant à l'ordre des points. Cela signifie que même si l'ordre des points dans le nuage varie, l'architecture doit être capable de traiter cette variation sans perte d'information.

Ainsi, **l'ordre des points dans un nuage n'a généralement pas de signification particulière** dans le cadre de PointNet, qui applique des opérations comme des permutations invariantes (par exemple, via des réseaux neuronaux sur des ensembles) pour garantir que la structure spatiale globale est préservée, indépendamment de l'ordre des points.
